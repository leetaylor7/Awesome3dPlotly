```python
#imports the dependancies 
import pandas as pd
import plotly as py
import plotly.graph_objs as go
```

```python
#Reads the data from a CSV
z_data = pd.read_excel('INSERT FILE PATH HERE', sheet_name = 'DESIRED SHEET NAME HERE')


#Sets up the chart data, including color scaling
data = [
    go.Surface(
        z=z_data.as_matrix(),
        cmin=850,
        cmax=950
    )
]


#This sets up the chart title and other layout options like the Axsis ticks ect...
layout = go.Layout(
    title='Rx 404 - Bed Profile',
    autosize=False,
    width=1000,
    height=1000,
    scene = dict(
        xaxis = dict(
            nticks=6,
            title='TI Location',
            range = [0,24],
        ),
        yaxis = dict(
            nticks=5, 
            title='Time',
            range = [0,495],
        ),
        zaxis = dict(
            nticks=5,
            title='Bed Temperature',
            range = [850,950],
        ),
    ),
    margin=dict(
        l=100,
        r=100,
        b=100,
        t=100
    )
)

#This actually produces the charts
fig = go.Figure(data=data, layout=layout, skip_invalid=True)
py.offline.plot(fig, filename='INSERT FILE NAME HERE')
```
